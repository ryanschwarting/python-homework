```python
# importing 
import csv
import os

# path to the csv file
df = os.path.join("..","PyBank","budget_data.csv")

# Initialize the metric variables
months = []
profit_changes = []

count = 0
total_profit = 0
last_month_pl = 0
this_month_pl = 0
profit_change = 0

# open and read the csv file
with open (df, newline="") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter =',')
    cvs_header = next(csv_file)

#use a for loop
    for row in csv_reader:
        
        # month count
        count += 1
        
         # calculate net total profit loss
        this_month_pl = int(row[1])
        total_profit += this_month_pl
        
        if count == 1:
            
            # last month profits is equal to this most recent month profit
            last_month_pl = this_month_pl
            continue 
        else:
            #cacluate change in profit
            profit_change = this_month_pl - last_month_pl
            
             # add each month to months list using the append feature
            months.append(row[0])
            
            # add profit change to the profit change list using the append feature
            profit_changes.append(profit_change)
            
            #set this this_month_profits to last_months_profit
            last_month_pl = this_month_pl
        
    # calculate the sum and average of the changes in profit/losses over the enitre period
    sum_profit = sum(profit_changes)
    average_profit = round(sum_profit/(count - 1), 2)
    
    # caclulate the greatest increase in profits over the entire period
    greatest_increase_pl = max(profit_changes)
    greatest_decrease_pl = min(profit_changes)
    
    # find the index of the greatest increase/decrease in profit losses 
    greatest_increase_month_index = profit_changes.index(greatest_increase_pl)
    greatest_decrease_month_index = profit_changes.index(greatest_decrease_pl)
    
    # assign the greatest increase and decrease month
    greatest_increase_month = months[greatest_increase_month_index]
    greatest_decrease_month = months[greatest_decrease_month_index]
    
# Print financial analysis
print('Financial Analysis')
print('----------------------------')
print(f"Total Months: {count}")
print(f"Total: {total_profit}")
print(f"Average Change: ${average_profit}")
print(f"Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase_pl})")
print(F"Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease_pl})")

# use the with statement in write mode to push a text file that prints the financial analysis
with open ('financial_analysis.txt', 'w') as financial:
    financial.write('Financial Analysis\n')
    financial.write('----------------------------\n')
    financial.write(f"Total Months: {count}\n")
    financial.write(f"Total: {total_profit}\n")
    financial.write(f"Average Change: ${average_profit}\n")
    financial.write(f"Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase_pl})\n")
    financial.write(F"Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease_pl})\n")
```
